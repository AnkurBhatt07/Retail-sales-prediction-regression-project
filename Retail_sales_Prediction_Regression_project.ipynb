Promo2Since[Year/Week] - describes the year and calendar week when the store started participating in Promo2

PromoInterval - describes the consecutive intervals Promo2 is started, naming the months the promotion is started anew. E.g. "Feb,May,Aug,Nov" means each round starts in February, May, August, November of any given year for that store

```python

# Function to fill missing values with random choice based on frequency
def fill_missing_with_weighted_random(column):
    non_missing_values = new_df[column].dropna()
    frequencies = non_missing_values.value_counts(normalize=True)
    missing_indices = new_df[column].isnull()
    
    new_df.loc[missing_indices, column] = np.random.choice(
        frequencies.index,
        size=missing_indices.sum(),
        p=frequencies.values
    )

# Fill missing values for both columns
fill_missing_with_weighted_random('CompetitionOpenSinceMonth')
fill_missing_with_weighted_random('CompetitionOpenSinceYear')


```

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H0 : Sales is normally distributed

H1 : Sales is not normally distributed.

#### 2. Perform an appropriate statistical test.

```python
sns.distplot(new_df['Sales'])
```

```python
from scipy.stats import shapiro

stat , p = shapiro(new_df['Sales'])
print(stat , p)
if p > 0.05:
    print(f'Failed to reject the Null Hypothesis with p = {p}')
else:
    print(f'Reject the Null Hypothesis with p = {p} . The sales is not normally distributed')

print('The significance level considered is 5% . If the probability is less than 5% than the data cannot be normally distributed.')
```

##### Which statistical test have you done to obtain P-Value?

We have conducted the shapiro wilk test to check for normality of Sales column .

##### Why did you choose the specific statistical test?

This test is used for checking the normality of data . If pval is less than the significance level , than we Reject the Null hypothesis and considers that the data is not normally distributed.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H0: The Customers and Sales have equal variances.

H1: Customers and sales do not have equal variances.

#### 2. Perform an appropriate statistical test.

```python
from scipy.stats import levene

stat , pval = levene(new_df['Customers'],new_df['Sales'])
print('stat = ' , stat)
print('pval',pval)
if pval > 0.05:
    print('Failed to reject the null hypothesis.The variances are equal ')
else:
    print('Rejected the Null Hypothesis.The variance of Sales and Customers are not equal')
```

##### Which statistical test have you done to obtain P-Value?

```python
# We have used the levene test to check the homoscedasticity of each independent variable with dependent variable.
```

##### Why did you choose the specific statistical test?

This test provides the pval less than 0.05 ie the significance level when the assumption of homoscedasticity is not met.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H0 : There is no correlation between year and sales.

H1 : There is correlation between year and sales.

```python

from scipy.stats import pearsonr
corr, pval = pearsonr(new_df['year'] , new_df['Sales'])
print(f'Pearson correlation coefficient: {corr}')
print(f'p-value: {pval}')
alpha = 0.05
if pval < alpha:
    print(f'Reject the Null Hypothesis.There is significant correlation between year and sales with p-value = {pval}')
else:
    print('Failed to reject the null hypothesis.There is not significant correlation betweeen Year and Sales')
```
